```python
def get_backdrop(soup):
    a = soup.findAll("div", {"class": "bq-box"})[0].span.text.replace("\n","").replace("\t","")
    b = soup.findAll("div", {"class": "bq-box"})[0].span.span
    if(b == None):
        return a
    else:
        b = b.text.replace("\n","").replace("\t","")
        return a[:len(a) - len(b)]
```

```python
get_backdrop(soup_lufax)
```

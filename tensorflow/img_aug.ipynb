```python
import imgaug,imageio
import imgaug as ia
from imgaug import augmenters as iaa
```

```python
image = imageio.imread("/home/qinjianbo/DATA/jianzipu_data/train/1/1_1_1577004744000.png")
```

```python
seq = iaa.Sequential([# 定义一个sequential，把要进行的图片操作（3个操作）放在里面
    iaa.Affine(rotate=(-25, 25)),
    iaa.AdditiveGaussianNoise(scale=(10, 30)),
    iaa.Crop(percent=(0, 0.4))
], random_order=True)
```

```python
images = [seq.augment_image(image) for i in range(3)]
```

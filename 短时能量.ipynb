```python
win_size = 441
def square_sum(frame):
    return np.sum(frame.T * frame)
```

```python
square_sum(meihua[442:883])
```

```python
sum_list = []
```

```python
index = 0
```

```python
while (index * 441 < len(meihua)):
    frame = meihua[index*441:(index+1)*441]
    sum = square_sum(frame)
    sum_list.append(sum)
    index +=1
```

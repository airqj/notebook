```python
def preprocess(data):
    return data - data[0][0]
    
def construct_map(path):
    point_map = {}
    path_student = path[0]
    path_teacher = path[1]
    for index in range(len(path_teacher)):
        point_teacher = path_teacher[index]
        point_student = path_student[index]
        if(point_map.get(point_teacher)):
            point_map[point_teacher].append(point_student)
        else:
            point_map[point_teacher] = [point_student]
    return point_map

def get_index_reduntant(distance_map):
    #print(distance_map)
    index_reduntant = sorted(distance_map.items(),key=lambda x:x[1])[0][0]
    return index_reduntant

def cmp_distance(points,reduce_indexes):
    data_student = points[0]
    data_teacher = points[1]
    distance_map = {}
    for index in reduce_indexes:
        d,_,_,_ = dtw(np.delete(data_student,index),data_teacher, dist=euclidean)
        distance_map[index] = d
    return distance_map
    
def reduce_point_map(points,point_map):
    result = {}
    for k,v in point_map.items():
        if(len(v) == 1):
            result[k] = v[0]
        else:
            distance_map = cmp_distance(points,v)
            index_reduntant = get_index_reduntant(distance_map)
            v.remove(index_reduntant)
            #print(index_reduntant)
            result[k] = v[0]
    return result

def get_point_map(data_student,data_teacher):
    s = preprocess(data_student)
    t = preprocess(data_teacher)
    d, cost_matrix, acc_cost_matrix, path = dtw(s, t, dist=euclidean)
    point_map = construct_map(path)
    result_reduce = reduce_point_map((s, t), point_map)

    return result_reduce

def get_illustration(points_map):
    illustration = ""
    illustration_format = "{0}\n"
    threshold = 1
    for point_teacher,point_student in points_map.items():
        if(abs(point_teacher - point_student) == threshold):
            illustration += illustration_format.format(point_teacher)
            threshold +=1
    return illustration

def parse_invalid_index(index_invalid):
    invalid_index_format = "{0}\n"
    result =""
    for index in index_invalid:
        result += invalid_index_format.format(index)
    return result

def filter_points(points,threshold):
    points_valid = []
    index_invalid = []
    for index in range(len(points)):
        if(points[index] > threshold):
            points_valid.append(points[index])
        else:
            index_invalid.append(index)
    return sum(points_valid) / len(points_valid),index_invalid

def cal_point(points_map,data_teacher,data_student):
    points = []
    points_map_items = list(points_map.items())
    indexes_previous = points_map_items[0]
    for index in range(1,len(points_map_items)):
        
        index_teacher = points_map_items[index][0]
        index_student = points_map_items[index][1]
        
        previous_index_student = indexes_previous[1]
        previous_index_teacher = indexes_previous[0]
        
        
        diff_student = abs(data_student[index_student][0] - data_student[previous_index_student][0])
        diff_teacher = abs(data_teacher[index_teacher][0] - data_teacher[previous_index_teacher][0])
        
        point = (1-(abs(diff_student - diff_teacher) / diff_teacher)) * 100
        points.append(point)
        
        indexes_previous = points_map_items[index]
#     return sum(points) / len(points)
    point,index_invalid = filter_points(points,0.3)
    illustration_invalid_index = parse_invalid_index(index_invalid)
    
    return point,illustration_invalid_index

def compute_point(data_student,data_teacher):
    points_map = get_point_map(data_student,data_teacher)
    illustration_reduntant = get_illustration(points_map)
    point,illustration_invalid = cal_point(points_map,preprocess(data_teacher),preprocess(data_student))
    
    return point,illustration_reduntant + illustration_invalid
```

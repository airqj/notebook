```python
import os,math
from pydub import AudioSegment
from pydub.playback import play
```

```python
length_per_slice = 20
output_dir = "/home/qinjianbo/DATA/fusui_spec/"
counter = 0
fusui_audio = "/home/qinjianbo/DATA/fusui_audio/"
```

```python
fileName = "张子谦-平沙落雁.mp3"
audio = AudioSegment.from_file("/home/qinjianbo/Music/张子谦/" + fileName)
audio = audio.set_channels(1)
```

```python
for item in os.listdir(fusui_audio):
    print(item)
```

```python
for fileName in os.listdir(fusui_audio):
    audio = AudioSegment.from_file(fusui_audio + fileName)
    audio = audio.set_channels(1)
    counter = 0
    length = math.ceil(audio.duration_seconds / length_per_slice)
    while (counter < length):
        start = length_per_slice * counter * 1000
        end   = length_per_slice * (counter + 1) * 1000
        slice = audio[start:end]
        audioFileName = output_dir + fileName.split(".")[0] + "_" + str(counter) + ".wav"
        specFileName  = output_dir + fileName.split(".")[0] + "_" + str(counter) + ".png"
        specCmd = "sox {0} -n spectrogram -x 1770 -y 380 -r -o {1}".format(audioFileName,specFileName)
        slice.export(audioFileName,format="wav")
        os.system(specCmd)
        print(specCmd)
        counter +=1
```
